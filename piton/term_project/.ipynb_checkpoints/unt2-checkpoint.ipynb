```python
import matplotlib.pyplot as plt

def draw_plot(X,y):
    for i in range(0,X.shape[1]):
        x = X[:,i]
        fig = plt.figure()
        plt.xlabel('X')
        plt.ylabel('y')
        plt.suptitle('X')
        plt.scatter(x, y, alpha=1,marker='.')
        fig.savefig(str(i)+'.png', dpi=100)
```
