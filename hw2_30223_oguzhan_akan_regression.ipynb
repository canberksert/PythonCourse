```python
def manual_regression(path='data.csv',target='target'):
    # Step 0. Import libraries and read the data
    def read_data(path, target):
        import pandas as pd
        try:
            df = pd.read_csv(path,sep=';')
            return df
        except TypeError:
            print('Data import unsuccessful, please make sure you have your data in csv format.')
        

    # Step 1. Normalisation
    def normalise(df):
        for col in df.columns:
            if(col!=target):
                df[col]=(df[col]-df[col].mean())/df[col].std()
        return df

    # Step 2. Listwise deletion of Null values
    def handle_na(df):
        na_rows = []
        for row in df.index:
            if(len(df.isna().iloc[row].unique())==2):
                na_rows.append(row)
        return df.drop(na_rows)

    df = read_data('data.csv','population')
    df = handle_na(normalise(df))
    return df
```

```python
manual_regression(path='data.csv',target='population')
```
